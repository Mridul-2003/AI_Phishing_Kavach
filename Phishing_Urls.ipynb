**Tokenization**

```python
from nltk.tokenize import word_tokenize
word_tokenize(data.URL[0],language ='english',preserve_line=False)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[A-Za-z]+')#to getting alpha only
# this will be pull letter which matches to expression
tokenizer.tokenize(data.URL[0]) # using first row
```

```python
data['tokenized_text']=data.URL.map(lambda t:tokenizer.tokenize(t))
data.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

data['text_stemmed'] = data['tokenized_text'].map(lambda l: [stemmer.stem(word) for word in l if word not in set(stopwords.words('english'))])
data.head()
```

```python
df = data.copy()
df.sample(10)
```

```python
df=df.drop(['text_stemmed'],axis=1)
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['lemmatized_text']=df['tokenized_text'].map(lambda l: [lemmatizer.lemmatize(word) for word in l if word not in set(stopwords.words('english'))])
df.sample(10)
```

```python
data['text_sent'] = data['text_stemmed'].map(lambda l: ' '.join(l))
data.sample(10)
```

```python
df['text_sent'] = df['lemmatized_text'].map(lambda l: ' '.join(l))
df.sample(10)
```

```python
data.URL[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer # create sparse matrix of words using regexptokenizes
cv = CountVectorizer()
feature = cv.fit_transform(data.text_sent) #transform all text which we tokenize and stemed
feature[:5].toarray()
```

```python
feature1 = cv.fit_transform(df.text_sent) #transform all text which we tokenize and stemed
feature1[:5].toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(feature,data.Label,test_size=0.2,stratify=data.Label)
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(feature1,df.Label,test_size=0.2,stratify=df.Label)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression()
lr1.fit(x_train1,y_train1)
```

```python
lr.score(x_test,y_test)
```

```python
lr1.score(x_test1,y_test1)
```

```python
from sklearn.pipeline import make_pipeline
pipeline_ls = make_pipeline(CountVectorizer(tokenizer = RegexpTokenizer(r'[A-Za-z]+').tokenize,stop_words='english'), LogisticRegression())
##(r'\b(?:http|ftp)s?://\S*\w|\w+|[^\w\s]+') ([a-zA-Z]+)([0-9]+)  -- these tolenizers giving me low accuray
```

```python
trainX, testX, trainY, testY = train_test_split(data.URL, data.Label)
```

```python
pipeline_ls.fit(trainX,trainY)
```

```python
pipeline_ls.score(testX,testY)
```

```python
prediction = pipeline_ls.predict(['https://www.linkedin.com'])
print(prediction[0])
```
